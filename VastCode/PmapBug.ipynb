## Importing the Packages

```python
from functools import partial

import jax.numpy as jnp
from jax import jit, pmap, random
```

```python
from jax.config import config
config.update("jax_enable_x64", True)
```

```python
numbermach = 8
```

```python
@partial(pmap, in_axes = 1, out_axes = 1)
def FFT(Array):
  return jnp.fft.rfft(Array, axis = -2)
```

```python
key = random.PRNGKey(0)
```

```python
tester = random.uniform(key, (5, 8, 10))
```

```python
testerfft = FFT(tester)
```

```python
print(testerfft.shape)
```

```python
@jit
def NestedFFT(Array):
  return FFT(Array)
```

```python
NestedFFT(tester)
```
